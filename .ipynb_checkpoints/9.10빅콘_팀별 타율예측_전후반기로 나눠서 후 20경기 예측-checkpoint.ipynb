## 2. train test 셋 만들기

```python
team=['HH', 'HT', 'KT', 'LG', 'LT', 'NC', 'OB', 'SK', 'SS', 'WO']
```

```python
def make_data(train,answer):
    train['HIT_RATE']=train['HIT']/train['AB']
    tr=train[['T_ID','HIT_RATE']]  #경기당 팀과 타율만 남김
    num=list(range(1,int((len(train)/10)+1)))*10  #팀별 경기순서 추가
    tr=tr.assign(order=num)
    tr=tr.pivot(index='T_ID',columns='order',values='HIT_RATE') #팀당 경기별 타율 데이터셋 만들기
    
    data=pd.merge(tr,answer,on='T_ID') #팀별 앞124경기와 후20경기 결과 합치기
    return data
```

```python
#정답 셋과 train 셋 만드는 함수
t_bat=[t_bat16,t_bat17,t_bat18,t_bat19]
def make_answer_train(year):
    a=t_bat[year-2016][['T_ID','AB','HIT']]
    
    train=pd.DataFrame()
    train_ans=pd.DataFrame()
    test=pd.DataFrame()
    test_ans=pd.DataFrame()
    
    for t in team:
        train=pd.concat([train,a[a['T_ID']==t][:52]])
        train_ans=pd.concat([train_ans,a[a['T_ID']==t][52:72]])
        test=pd.concat([test,a[a['T_ID']==t][72:124]])
        test_ans=pd.concat([test_ans,a[a['T_ID']==t][124:]])
        
    #팀별 AB,HIT합 구하기
    train_ans=train_ans.groupby(['T_ID']).sum()
    train_ans['HIT_RATE']=train_ans['HIT']/train_ans['AB'] #타율 계산
    train_ans=train_ans.drop(train_ans.columns[[0,1]],axis='columns')
    
    test_ans=test_ans.groupby(['T_ID']).sum()
    test_ans['HIT_RATE']=test_ans['HIT']/test_ans['AB'] #타율 계산
    test_ans=test_ans.drop(test_ans.columns[[0,1]],axis='columns')
    
    data1=make_data(train,train_ans)
    data2=make_data(test,test_ans)

    return data1,data2
```

```python
train16,test16=make_answer_train(2016)
train17,test17=make_answer_train(2017)
train18,test18=make_answer_train(2018)
train19,test19=make_answer_train(2019)
```

## 3. 모델학습

```python
# 모델 모듈
import xgboost
from xgboost import XGBRegressor, plot_importance
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.linear_model import Ridge, Lasso, LinearRegression, ElasticNet
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
train19
```

```python
X_cols=list(range(1,52))
Y_col=['HIT_RATE']
```

```python
#### 학습시킬 모델
model_dict = {'xgb':XGBRegressor(n_estimators=100, learning_rate=0.05 ,max_depth=3, random_state=23),
    'lgbm':LGBMRegressor(),
    'rf':RandomForestRegressor(),
    'svr':SVR(),
    'reg':LinearRegression(),
    'ridge':Ridge(),
    'lasso':Lasso()}
```

```python
train19
```

```python
train=train19
test=test19
```

```python
X_train = train[X_cols]
Y_train = train[Y_col]

X_test = test[X_cols]
Y_test = test[Y_col]

model = model_dict['svr']    #모델명 바꾸어 가면서 실험 가능
#weight = train['17_HIT_RATE']
#model.fit(X_train, Y_train, sample_weight=weight)
model.fit(X_train, Y_train)  #가중치 제거
Y_predict = model.predict(X_test)

#test_error=mean_squared_error(Y_test, Y_predict, sample_weight=X_test['17_HIT_RATE']) ** 0.5
test_error=mean_squared_error(Y_test, Y_predict) ** 0.5  #가중치 제거
r2=r2_score(Y_test, Y_predict)
    
print("test error : ", test_error)
print("r2 : ", r2)
```
