```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1. 데이터 가져오기

```python
data=pd.read_csv("C:/Users/a/Desktop/2020빅콘/데이터/dataset.csv",encoding='utf-8',engine='python')
```

```python
#모든 열 다 출력하기
pd.options.display.max_columns=None
```

```python
data=data.drop(data.columns[[0]],axis='columns')
```

```python
#함수 만들기
def train_test(data,ratio=0.7):
    for i in range(len(data)):
        train=data.sample(frac=ratio,random_state=2020)
        test=data.drop(train.index)
    return train,test
```

```python
train,test=train_test(data)
```

```python
train=train.set_index('T_ID')
test=test.set_index('T_ID')
test
```

## 3. 모델학습

```python
# 출력 할 때 마다, 기존 출력물들은 제거해주는 모듈
from IPython.display import clear_output
```

```python
# 모델 모듈
import xgboost
from xgboost import XGBRegressor, plot_importance
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.linear_model import Ridge, Lasso, LinearRegression, ElasticNet
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
X_cols=['HITRATE', 'OBS', 'SLG', 'ERA','rel_HITRATE', 'rel_ERA']
Y_col=['AFTER_20']
```

```python
#### 학습시킬 모델
model_dict = {'xgb':XGBRegressor(n_estimators=100, learning_rate=0.05 ,max_depth=3, random_state=23),
    'lgbm':LGBMRegressor(),
    'rf':RandomForestRegressor(),
    'svr':SVR(),
    'reg':LinearRegression(),
    'ridge':Ridge(alpha=100,normalize=True),
    'lasso':Lasso(alpha=0.05,normalize=True)}  #변수 자동 채택 / 많은 변수를 다룰때 활용
```

*ridge : 모든 변수를 가지고 계수값을 줄임(모델 복잡/성능저하)

*lasso : 변수들끼리 correlate하면 한개의 변수만 채택(정보손실 / 정확성 떨어질수있음)

```python
X_train = train[X_cols]; Y_train = train[Y_col]
X_test = test[X_cols]; Y_test = test[Y_col]
diff=[]

#모델 여러개 돌려보기
model_list=['xgb','lgbm','rf','svr','reg','ridge','lasso']
for m in model_list:
    model = model_dict[m]
    if m=='rf': Y_train=Y_train.values.ravel()  #rf에서의 오류 없애기 위함
    model.fit(X_train, Y_train)
    Y_predict = model.predict(X_test)

    r2=r2_score(Y_test, Y_predict)
    print(m,"\t>  r2 :", r2,'\t훈련 세트 점수 :{:.3f}'.format(model.score(X_train,Y_train)))
    diff.append(r2)
```

```python
#r2가 가장 높은 모델 돌리기
model = model_dict[model_list[diff.index(max(diff))]]
if m=='rf': Y_train=Y_train.values.ravel()  #rf에서의 오류 없애기 위함
model.fit(X_train, Y_train)
Y_predict = model.predict(X_test)
print(model)
```

### 예측값과 실제값 차이

```python
# 예측값과 실제값의 시각화
plt.scatter(Y_test, Y_predict)
plt.xlabel("실제 값")
plt.ylabel("예측 값")
```

```python
#예측값과 실제값 차이
a=Y_test
a=a.assign(predict=Y_predict)
a['diff']=a['AFTER_20']-a['predict']
a
```

```python
#팀별 타율 평균
b=a.filter(['AFTER_20','predict'])
b=b.groupby(['T_ID']).mean()
b['diff']=b['AFTER_20']-b['predict']
b
```

```python
#차이 합
abs(b['diff']).sum()
```
