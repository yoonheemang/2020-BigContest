## 2. 월 정보 추가

```python
#월 정보 추가
t_bat19['MONTH']=t_bat19['GDAY_DS'].astype(str).str[4:6]
```

```python
stats=['PA','AB','RBI','RUN','HIT','H1','H2','H3','HR','SB','CS','SH','SF','BB','IB','HP'
       ,'KK','GD','ERR','LOB','P_AB_CN','P_HIT_CN']
```

## 3. 각 팀의 월별 타율 변수로 생성

```python
#팀별 기록(19년도)
data19=t_bat19.filter(['T_ID','MONTH','AB','HIT'])
t19=data19.groupby(['T_ID','MONTH']).sum()
t19['HIT_RATE']=t19['HIT']/t19['AB']
t19
```

```python
t19=t19.drop(t19.columns[[0,1]],axis='columns')
t19.head()
```

```python
col = [i for i in 10 + 'HIT_RATE']
col
```

```python
t19=t19.reset_index(level=['MONTH']) #MONTH열을 index에서 값으로 변환
t19['MONTH']
```

```python
t19=t19.reset_index(level=['MONTH']) #MONTH열을 index에서 값으로 변환
t19=t19.astype({'MONTH':'int'}) #MONTH 값 형식을 숫자로 변환
```

```python
t19_8_1 = t19[t19['MONTH']<=8] 
result1 = t19[t19['MONTH']>8]

t19_8=t19_8_1.groupby(['T_ID']).sum()
result1=result1.filter(['19.8_T_ID','19.8_AB','19.8_HIT'])
result=result1.groupby(['T_ID']).sum()

t19_8=t19_8.drop(t19_8.columns[[0]],axis='columns')
```

```python
t19_8['19.8_HIT_RATE']=t19_8['19.8_HIT']/t19_8['19.8_AB']
result['result_HIT_RATE']=result['19.8_HIT']/result['19.8_AB']
result=result.drop(result.columns[[0,1]],axis='columns')
```

## 6. 학습시킬 데이터 셋 만들기

```python
#outer joing을 이용하여 데이터 합치기
a=pd.merge(t16,t17,on=['T_ID'],how='outer')
b=pd.merge(a,t18,on=['T_ID'],how='outer')
c=pd.merge(b,t19_8,on=['T_ID'],how='outer')
data=pd.merge(c,result,on=['T_ID'],how='outer')
```

## 7. 모델학습

```python
# 출력 할 때 마다, 기존 출력물들은 제거해주는 모듈
from IPython.display import clear_output
```

```python
# 모델 모듈
import xgboost
from xgboost import XGBRegressor, plot_importance
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.linear_model import Ridge, Lasso, LinearRegression, ElasticNet
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
stats+=['HIT_RATE']
```

```python
# 누적 및 lag stat features 이름
stats_16 = ['16_' + i for i in stats]
stats_17 = ['17_' +i for i in stats]
stats_18 = ['18_' +i for i in stats]
stats_19 = ['19.8_' +i for i in stats]
```

```python
X_cols=stats_16+stats_17+stats_18
Y_col=['result_HIT_RATE'] #19_AB, 19_HIT
```

```python
#### 학습시킬 모델
model_dict = {'xgb':XGBRegressor(n_estimators=110, learning_rate=0.05 ,max_depth=3, min_samples_split=400, random_state=23),
    'lgbm':LGBMRegressor(),
    'rf':RandomForestRegressor(),
    'svr':SVR(),
    'reg':LinearRegression(),
    'ridge':Ridge(),
    'lasso':Lasso()}
```

```python
train=data
test=data
```

```python
X_train = train[X_cols]
Y_train = train['19.8_HIT_RATE']

X_test = test[X_cols]
Y_test = test[Y_col]

model = model_dict['rf']    #모델명 바꾸어 가면서 실험 가능
weight = train['17_HIT_RATE']
model.fit(X_train, Y_train, sample_weight=weight)
Y_predict = model.predict(X_test)

test_error=mean_squared_error(Y_test, Y_predict, sample_weight=X_test['17_HIT_RATE']) ** 0.5
r2=r2_score(Y_test, Y_predict)
 
    
print("test error : ", test_error)
print("r2 : ", np.mean(r2))
print("완료")
```
