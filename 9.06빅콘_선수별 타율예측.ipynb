```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1. 데이터 가져오기

```python
p16=pd.read_csv("C:/Users/a/Desktop/2020빅콘/데이터/data2016.csv",engine='python')
p17=pd.read_csv("C:/Users/a/Desktop/2020빅콘/데이터/data2017.csv",engine='python')
p18=pd.read_csv("C:/Users/a/Desktop/2020빅콘/데이터/data2018.csv",engine='python')
p19=pd.read_csv("C:/Users/a/Desktop/2020빅콘/데이터/data2019.csv",engine='python')
```

```python
#모든 열 다 출력하기
pd.options.display.max_columns=None
```

```python
stats=['START_CK', 'AB', 'RBI', 'RUN', 'HIT', 'H2', 'H3', 'HR', 'SB',
       'CS', 'SH', 'SF', 'BB', 'IB', 'HP', 'KK', 'GD', 'ERR', 'LOB', 'P_AB_CN',
       'P_HIT_CN', 'HR_CN', 'BABIP', 'HIT_RATE', 'DR', 'SSG', 'H1', 'SLG',
       'OBP', 'OPS','AGE_VA','WON']
```

## 2. 데이터 살펴보기

```python
#P_ID index로 변환
p16=p16.set_index('P_ID')
p17=p17.set_index('P_ID')
p18=p18.set_index('P_ID')
p19=p19.set_index('P_ID')
```

```python
#AB가 10보다 작으면 지우기
p16=p16[p16['AB']>10]
p17=p17[p17['AB']>10]
p18=p18[p18['AB']>10]
p19=p19[p19['AB']>0]
```

```python
p16=p16.add_prefix('16_') #각 변수명 앞에 연도 붙이기
p17=p17.add_prefix('17_') #각 변수명 앞에 연도 붙이기
p18=p18.add_prefix('18_') #각 변수명 앞에 연도 붙이기
```

## 3. 학습시킬 데이터 셋 만들기

```python
data19=p19.filter(['P_ID','AB','HIT'])
p19=data19.groupby(['P_ID']).sum()
p19['HIT_RATE']=p19['HIT']/p19['AB']
p19=p19.add_prefix('19_')
p19.head()
```

### 3-1. outer join을 한 후 NAN값 0으로 대체

0으로 대체할 경우 inner join보다 설명력이 낮게 나옴.

NAN값을 다른 연도의 값으로 대체해볼것

```python
#팀, 상대팀, 월 을 기분으로 데이터 합치기
a=pd.merge(p16,p17,on=['P_ID'],how='outer')
b=pd.merge(a,p18,on=['P_ID'],how='outer')
outer_data=pd.merge(b,p19,on=['P_ID'],how='inner')  #결과값은 무조건 있어야되므로...
```

```python
#결측값 0으로 처리
outer_data=outer_data.fillna(0)
```

```python
outer_data
```

### 3-2. inner join 하기

```python
a=pd.merge(p16,p17,on=['P_ID'],how='inner')
b=pd.merge(a,p18,on=['P_ID'],how='inner')
inner_data=pd.merge(b,p19,on=['P_ID'],how='inner')
```

```python
inner_data.head()
```

## 4. 모델학습

```python
# 출력 할 때 마다, 기존 출력물들은 제거해주는 모듈
from IPython.display import clear_output
```

```python
# 모델 모듈
import xgboost
from xgboost import XGBRegressor, plot_importance
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.linear_model import Ridge, Lasso, LinearRegression, ElasticNet
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# 누적 및 lag stat features 이름
stats_16 = ['16_' + i for i in stats]
stats_17 = ['17_' +i for i in stats]
stats_18 = ['18_' +i for i in stats]
```

```python
X_cols=stats_16+stats_17+stats_18
#타율 예측
Y1_col=['19_HIT_RATE']
```

```python
#타수 예측
Y2_col=['19_AB']
#안타 예측
Y3_col=['19_HIT']
```

```python
#### 학습시킬 모델
model_dict = {'xgb':XGBRegressor(n_estimators=50, learning_rate=0.05 ,max_depth=3, random_state=23),
    'lgbm':LGBMRegressor(),
    'rf':RandomForestRegressor(),
    'svr':SVR(),
    'reg':LinearRegression(),
    'ridge':Ridge(),
    'lasso':Lasso()}
```

train test 만들기

```python
#함수 만들기
def train_test(data,ratio=0.7):
    for i in range(len(data)):
        train=data.sample(frac=ratio,random_state=2020)
        test=data.drop(train.index)
    return train,test
```

```python
#outer인 경우
train,test=train_test(outer_data)
```

```python
#inner인 경우
train,testt=train_test(inner_data)
```

### 타율 예측

```python
X_train = train[X_cols]
Y1_train = train[Y1_col]

X_test = test[X_cols]
Y1_test = test[Y1_col]

model = model_dict['lgbm']    #모델명 바꾸어 가면서 실험 가능
weight = train['18_AB']
model.fit(X_train, Y1_train, sample_weight=weight)
#model.fit(X_train, Y_train)
Y1_predict = model.predict(X_test)

test_error=mean_squared_error(Y1_test, Y1_predict, sample_weight=X_test['18_AB']) ** 0.5
#test_error=mean_squared_error(Y_test, Y_predict) ** 0.5

r2=r2_score(Y1_test, Y1_predict)
 
    
print("test error : ", test_error)
print("r2 : ", r2)
```

### 타수 예측

```python
X_train = train[X_cols]
Y2_train = train[Y2_col]

X_test = test[X_cols]
Y2_test = test[Y2_col]

model = model_dict['xgb']    #모델명 바꾸어 가면서 실험 가능
weight = train['18_AB']
model.fit(X_train, Y2_train, sample_weight=weight)
Y2_predict = model.predict(X_test)

test_error=mean_squared_error(Y2_test, Y2_predict, sample_weight=X_test['18_AB']) ** 0.5
r2=r2_score(Y2_test, Y2_predict)
 
    
print("test error : ", test_error)
print("r2 : ", r2)
```

### 안타 예측

```python
X_train = train[X_cols]
Y3_train = train[Y3_col]

X_test = test[X_cols]
Y3_test = test[Y3_col]

model = model_dict['xgb']    #모델명 바꾸어 가면서 실험 가능
weight = train['18_AB']
model.fit(X_train, Y3_train, sample_weight=weight)
Y3_predict = model.predict(X_test)

test_error=mean_squared_error(Y3_test, Y3_predict, sample_weight=X_test['18_AB']) ** 0.5
r2=r2_score(Y3_test, Y3_predict)
 
    
print("test error : ", test_error)
print("r2 : ", r2)
```
