## 4. 변수 생성하기

```python
stats=['START_CK', 'AB', 'RBI', 'RUN', 'HIT', 'H2', 'H3', 'HR', 'SB',
       'CS', 'SH', 'SF', 'BB', 'IB', 'HP', 'KK', 'GD', 'ERR', 'LOB', 'P_AB_CN',
       'P_HIT_CN', 'HR_CN', 'BABIP', 'HIT_RATE', 'DR', 'SSG', 'H1', 'SLG',
       'OBP', 'OPS','AGE_VA','WON']
```

```python
def make_var(data):
    data=data.drop(data.columns[[0,1,23]],axis='columns')
    data=data.rename({'PCODE':'P_ID'},axis='columns')
    data=data.set_index('P_ID')
    data['WON']=pd.to_numeric(data['WON'])
    data['HR_CN']=data['P_HIT_CN']/data['P_AB_CN'] #득점권 타율변수
    data['BABIP']=(data['HIT']-data['HR'])/(data['AB']-data['KK']-data['HR']+data['SF']) #BABIP변수
    data['HIT_RATE'] = data['HIT']/data['AB']     #타율 변수 만들기
    data['DR'] = data['SB']+data['CS']            #도루시도 변수 만들기
    data['SSG'] = data['BB']+data['IB']+data['HP']             #사사구 변수 만들기
    data['H1'] = data['HIT']-data['H2']-data['H3']-data['HR']   #1루타 변수 만들기
    data['SLG'] = (data['H1']+data['H2']*2+data['H3']*3+data['HR']*4)/(data['AB'])       #장타율 변수 만들기
    data['OBP'] = (data['HIT']+data['BB']+data['HP'])/(data['AB']+data['BB']+data['HP']+data['SF']) #출루율 변수 만들기
    data['OPS'] = data['SLG']+data['OBP']         #OPS 변수 만들기

    #결측값 0으로 처리
    data=data.fillna(0)
    return data
```

```python
train16=make_var(train_16)
test16=make_var(test_16)
train17=make_var(train_17)
test17=make_var(test_17)
train18=make_var(train_18)
test18=make_var(test_18)
train19=make_var(train_19)
test19=make_var(test_19)
```
