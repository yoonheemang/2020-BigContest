## 3. train / test set 나누기

```python
#함수 만들기
def train_test(data,ratio=0.7):
    for i in range(len(data)):
        train=data.sample(frac=ratio,random_state=2020)
        test=data.drop(train.index)
    return train,test
```

```python
#train test 데이터 셋 이름 만들기
datalist=[i + str(j) for j in range(16,20) for i in ['first','second']] #first16, second16, first17 ...
tt_data_list = [i + '_' + j for i in datalist for j in ['train','test']] #first16_train, first16_test, second16_train
```

```python
first16_train,first16_test=train_test(first16)
second16_train,second16_test=train_test(second16)
first17_train,first17_test=train_test(first17)
second17_train,second17_test=train_test(second17)
first18_train,first18_test=train_test(first18)
second18_train,second18_test=train_test(second18)
first19_train,first19_test=train_test(first19)
second19_train,second19_test=train_test(second19)
```

```python
#팀별 타율 생성 함수
def add_rate(dataset,year,filter_list=['T_ID','AB','HIT'],group_list=['T_ID']):
    a=dataset.filter(filter_list)
    b=a.groupby(group_list).sum()
    b['HIT_RATE']=b['HIT']/b['AB']
    b=b.add_prefix(year)
    return b
```

```python
#train data
fir16_train=add_rate(first16_train,'16_first_')   #16전반기
sec16_train=add_rate(second16_train,'16_second_') #16후반기
fir17_train=add_rate(first17_train,'17_first_')   #17전반기
sec17_train=add_rate(second17_train,'17_second_') #17후반기
fir18_train=add_rate(first18_train,'18_first_')   #18전반기
sec18_train=add_rate(second18_train,'18_second_') #18후반기
fir19_train=add_rate(first19_train,'19_first_')   #19전반기
sec19_train=add_rate(second19_train,'19_second_') #19후반기 -> 예측 대상
```

```python
train=pd.concat([fir16_train,sec16_train,fir17_train,sec17_train,fir18_train,sec18_train,fir19_train,sec19_train],axis=1)
train.head()
```

```python
#test data
fir16_test=add_rate(first16_test,'16_first_')   #16전반기
sec16_test=add_rate(second16_test,'16_second_') #16후반기
fir17_test=add_rate(first17_test,'17_first_')   #17전반기
sec17_test=add_rate(second17_test,'17_second_') #17후반기
fir18_test=add_rate(first18_test,'18_first_')   #18전반기
sec18_test=add_rate(second18_test,'18_second_') #18후반기
fir19_test=add_rate(first19_test,'19_first_')   #19전반기
sec19_test=add_rate(second19_test,'19_second_') #19후반기 -> 예측 대상
```

```python
test=pd.concat([fir16_test,sec16_test,fir17_test,sec17_test,fir18_test,sec18_test,fir19_test,sec19_test],axis=1)
test.head()
```

```python
#### 학습시킬 모델
model_dict = {'xgb':XGBRegressor(n_estimators=110, learning_rate=0.05 ,max_depth=3, min_samples_split=400, random_state=23),
    'lgbm':LGBMRegressor(),
    'rf':RandomForestRegressor(),
    'svr':SVR(),
    'reg':LinearRegression(),
    'ridge':Ridge(),
    'lasso':Lasso()}
```
